```python
strings = ['http', 'https', 'www', '.', 'com', ':', '//', '/']
url_names = []

for url in list(df['URLs']):
    for string in strings:
        string in url: 
            new_url = url.replace(string, '')
            if string not in url: 
                url_names.append(new_url)
```
