```python
import re
import requests
from bs4 import BeautifulSoup
import pandas as pd
import numpy as np
```

```python
url = 'https://www.consob.it/web/consob-and-its-activities/warnings?viewId=ultime_com_tutela'
response = requests.get(url, headers={"Accept-Language":"en-US"})
soup = BeautifulSoup(response.content, "html.parser")
```

```python
soup.find_all('div', class_='documentContent consobContent')
```

```python
for i in soup.find_all('div', class_='documentContent consobContent'):
    regulator = i.find_all('u')
    print(regulator)
```

```python
soup.find_all('li', class_='text').text
```

```python
all_text = []
for i in soup.find_all('li', class_='text'):
    regulator = i.find('p').text.strip()
    all_text.append(regulator)
```

```python
dummy_lst=all_text[4].split()
```

```python
urls_dummy=[]
for word in dummy_lst:
    if word.startswith('https') or word.startswith('(https'): #or 'https' in word:
        urls_dummy.append(word)
```

```python
split_text = [x.split() for x in all_text]
```

```python
all_urls = []
for lst in split_text:
    for word in lst:
        if word.startswith('https') or word.startswith('(https') or word.startswith('www') or word.startswith('(www'): #or 'https' in word:
            all_urls.append(word)
```

```python
len(all_urls)
```

```python
df = pd.DataFrame(np.array(all_urls))
```

```python
df_new = df[0].str.strip('(').str.strip(')').str.strip(',')
```

```python
df_new
```

```python
new_urls = []
for url in df_new:
    if not 'http' in url: 
        new_url = 'http://' + url
        new_urls.append(new_url)
    else:
        new_urls.append(url)
        
```

```python
df_clean_urls = pd.DataFrame(np.array(new_urls))
```

```python
df_clean_urls.to_csv('more_scam_URLs_0212.csv')
```
